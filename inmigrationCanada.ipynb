```python
df_canada = df_canada.set_index("Country")
df_canada
```

```python
df_canada['Total'] = df_canada.iloc[:, 8:42].sum(axis=1)
df_canada
```

```python
df_canada.dtypes
```

```python
years = list(map(int, range(1980, 2014)))


df_canada.loc["Haiti", years].plot(kind = 'line')

plt.title('Inmigration from Haiti')
plt.ylabel('Number of inmigrants')
plt.xlabel('years')
```

```python
df_canada.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_canada
```

```python
df_top5 = df_canada.head(5)
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.loc[years].plot(kind = 'area')

plt.title('Inmigration trend of Top5 Countries')
plt.ylabel('Number of inmigrants')
plt.xlabel('Years')
```
